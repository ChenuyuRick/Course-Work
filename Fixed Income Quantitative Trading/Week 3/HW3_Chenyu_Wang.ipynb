## Q1

From the question, we can get the information that:
$$
dr(t) = {\mu - \kappa r(t)}dt + \sqrt{\gamma r(t) + \sigma}dW(t)
$$
Now, Let's suppose that the price of P can be written as:
$$
p(t,T) = e^{A(t,T) - B(t, T)r(t)}
$$
Apply Ito's Lemma:
$$
dp(t,T) = \frac{\partial p(t,T)}{\partial t}dt + \frac{\partial p(t,T)}{\partial r}dr + \frac{1}{2}\frac{\partial^2 p(t,T)}{\partial r^2}(dr)^2
$$
Individually, we can sovle the partial derivatives and get the result:
$$
\frac{\partial p(t,T)}{\partial t} = (\frac{dA(t,T)}{dt} - r(t)\frac{dB(t,T)}{dt})p(t,T)
$$
$$
\frac{\partial p(t,T)}{\partial r} = -B(t,T)p(t,T)
$$
$$
\frac{\partial^2 p(t,T)}{\partial r^2} = B^2(t,T)p(t,T)
$$
Now, plug in the result above, we get that:
$$
\frac{dp(t,T)}{p(t,T)} = (\frac{dA(t,T)}{dt} - r(t)\frac{dB(t,T)}{dt})dt - B(t,T)[(\mu - \kappa r(t))dt + \sqrt{\gamma r(t) + \sigma}dW(t)]+ \frac{1}{2}B^2(t,T)(\gamma r(t) + \sigma)dt
$$
rearange the terms in the equation, we get:
$$
\frac{dp(t,T)}{p(t,T)} = [\frac{dA(t,T)}{dt}-r(t)\frac{dB(t,T)}{dt} - B(t,T)(\mu - \kappa r(t)) + \frac{1}{2}B^2(t,T)(\gamma r(t) + \sigma)]dt-B(t,T)\sqrt{\gamma r(t) + \sigma}dW(t)
$$
Under the risk-neutral measure, we have:
$$
\frac{dA(t,T)}{dt}-r(t)\frac{dB(t,T)}{dt} - B(t,T)(\mu - \kappa r(t)) + \frac{1}{2}B^2(t,T)(\gamma r(t) + \sigma) = r
$$
Then we can get the following equations:
$$
\frac{dA(t,T)}{dt} - B(t,T)\mu + \frac{1}{2}B^2(t,T)\sigma = (1+\frac{dB(t,T)}{dt}-\kappa B(t,T)-\frac{1}{2}B^2(t,T)\gamma)r(t)
$$
In order to satisfy the equation above, we should satisfy the following two equations:
$$
\frac{dA(t,T)}{dt}-B(t,T)\mu + \frac{1}{2}\sigma B^2(t,T) = 0
$$
$$
\frac{dB(t,T)}{dt}-\kappa B(t,T)-\frac{1}{2}\gamma B^2(t,T) + 1 = 0
$$
End the proof

### Method 1 : Solving the ODE
#### Step 1 : Solve B
First, we solve B, according to the condition $\gamma=0$, the ODE turns to:
$$
\frac{dB(t,T)}{dt} -\kappa B + 1 = 0
$$
Rearrange the equation, we can get:
$$
\frac{dB(t,T)}{\kappa B - 1} = dt
$$
Integral both side of the equation we can get:
$$
B(t,T) = \frac{1}{\kappa}[1-C_1 e^{-\kappa (T-t)}]
$$
According to the model above, we need to find the constraints, when t = T, $p(T,T) = 1$ Therefore,
$$
p(T,T) = e^{A(T,T)-B(T,T)r(T)} = 1
$$
Since $r(T)$ is a stochastic variable and the equation above should always hold, so we have the constrains:
$$
A(T,T) = 0
$$
$$
B(T,T) = 0
$$
Plug in the constraints, we can get that $C_1 = 1$, plug in $\kappa = 0.05$, we can get B:
$$
B(t,T) = 20 - 20e^{-\frac{1}{20}(T-t)}
$$

#### Step 2: Solve A
Plug in B into A's ODE:
$$
dA(t,T) = [0.05 - 0.05e^{-0.05(T-t)}] - \frac{1}{200}[400-800e^{-0.05(T-t)}+400e^{-0.1(T-t)}]dt
$$
Solve this ODE with constraint $A(T,T) = 0$ and get A:
$$
A(t,T) = -59 + \frac{19}{20}(T-t) + 79e^{-0.05(T-t)}-20e^{-0.1(T-t)}
$$

#### Step 3: Plug in A and B, get P
$$
p(t,T)=e^{A(t,T)-B(t,T)r(t)}=e^{-59.2+1.95(T-t)-20e^{-0.1(T-t)}+79.2e^{-0.05(T-t)}}
$$

### Method 2 : Analytic Method

#### Step 1 : Solve the r(t) expression
First, we know that $\gamma = 0$, So the dynamics of r is:
$$
dr(t) = (\mu - \kappa r(t))dt + \sqrt{\sigma} dW(t)
$$
Multiply $e^{\kappa t}$ on both sides of the equation and rearrange it:
$$
e^{\kappa t}dr(t) + \kappa r(t)e^{\kappa t}dt = e^{\kappa t}\mu dt + \sqrt{\sigma}e^{\kappa t}dW(t)
$$
And now we can find that the left hand side of the equation is:
$$
e^{\kappa t}dr(t) + \kappa r(t)e^{\kappa t}dt = d(e^{\kappa t}r(t))
$$
So, we integral both sides of the equation and finally get the expression of r(t):
$$
r(t) = r(0)e^{-\kappa t} + \frac{\mu}{\kappa}[1-e^{-\kappa t}]+\sqrt{\sigma}\int_{0}^{t}e^{-\kappa(t-s)}dW_s
$$

#### Step 2 : Compute the expectation and variance  of z(t)
Suppose that:
$$
z(t,T) = -\int_{t}^{T}r(u)du
$$
Plug in the r(t) expresison that we got above:
$$
z(t,T) = -\frac{r(t)}{\kappa}[1-e^{-\kappa(T-t)}]-\frac{\mu}{\kappa}(T-t)-\frac{\mu}{\kappa^2}[e^{-\kappa (T-t)}-1]-\sqrt{\sigma}\int_{t}^{T}\int_{t}^{u}e^{-\kappa(u-s)}dW_sdu
$$
So we could solve that the expectation under Q measure is:
$$
E_{t,r}^{Q}[z(t,T)]=-\frac{r(t)}{\kappa}[1-e^{-\kappa(T-t)}]-\frac{\mu}{\kappa}(T-t)-\frac{\mu}{\kappa^2}[e^{-\kappa (T-t)}-1]
$$
And the variance under Q measure is:
$$
Var_{t,r}^{Q}[z(t,T)]=\sigma \int_{t}^{u}(\int_{t}^{T}e^{-\kappa(u-s)}du)^2ds = \frac{\sigma}{\kappa^2}[(T-t)+\frac{1}{2\kappa}(1-e^{-2\kappa(T-t)})-\frac{2}{\kappa}(1-e^{-\kappa(T-t)})]
$$

#### Step 3 : Compute the price of p(t,T)
We know that:
$$
p(t,T) = E_{t,r}^{Q}[e^{-\int_{t}^{T}r(u)du}] = E_{t,r}^{Q}[e^{z(t,T)}]
$$
Since Z(t,T) is normally distributed, 
$$
p(t,T) = e^{E_{t,r}^{Q}[z(t,T)] + \frac{1}{2}Var_{t,r}^{Q}[z(t,T)]}
$$
Plug in the result above and $\kappa = 0.05$, $\mu = 0.0025$, $\sigma = 0.01$, we get that:
$$
p(t,T)=e^{A(t,T)-B(t,T)r(t)}=e^{-59.2+1.95(T-t)-20e^{-0.1(T-t)}+79.2e^{-0.05(T-t)}}
$$
It's exactly the same formula as that was shown in Method 1. So the two methods provide the same result.

### Method 1 : LevelRegression

```python
import statsmodels.api as sm
from statsmodels.tsa.stattools import adfuller
```

```python
def LevelRegression(comb, data):
    Y = data[comb[0]]
    X = data[comb[1]]
    X = sm.add_constant(X)
    model = sm.OLS(Y,X).fit()
    return model.params[1]
```

#### Compute the P_value and see if the series is stationary

```python
Asset1 = []
Asset2 = []
Weight_list = []
insample_PValue = []
insample_St = []
outsample_PValue = []
outsample_St = []
for comb in comb_list:
    #Save comb
    Asset1.append(comb[0])
    Asset2.append(comb[1])
    
    #Compute Weight
    Weight = LevelRegression(comb, sample1)
    Weight_list.append(Weight)
    
    #Compute Portfolio
    Portfolio_in = sample1[comb[0]] - Weight*sample1[comb[1]]
    Portfolio_out = ousample[comb[0]] - Weight*ousample[comb[1]]
    
    #ADF Test
    ADF_in = adfuller(Portfolio_in)
    ADF_out = adfuller(Portfolio_out)
    
    #Stationarity
    insample_PValue.append(ADF_in[1])
    outsample_PValue.append(ADF_out[1])
    if ADF_in[1] < 0.05:
        insample_St.append('Yes')
    else:
        insample_St.append('No')
    
    if ADF_out[1] < 0.05:
        outsample_St.append('Yes')
    else:
        outsample_St.append('No')
    
    ##Combine the result
    di = {'Asset1':Asset1, 
          'Asset2':Asset2, 
          'Weight':Weight_list, 
          'Insample_P':insample_PValue, 
          'Stationarity_in':insample_St, 
          'Outsample_P':outsample_PValue,
          'Stationarity_out':outsample_St
         }
    result_LR = pd.DataFrame(di)
result_LR
```

### Method 2 : CCA_Chou_Ng

```python
def CCA_Chou_Ng(comb, data):
    
    data_set = data[comb]
    
    #data_set is pandas dataframe
    df_lag = data_set.shift(1).dropna()
    df = data_set.drop(data_set.index[0]).dropna()
    n = len(data_set.columns)
    
    #X(t) ~ M_1 + X(t-1)
    X = df_lag.as_matrix()
    X_I = sm.add_constant(X)
    Y = df.as_matrix()
    l1 = sm.OLS(Y,X_I).fit()
    B=l1.params[1:(n+1)]
    
    #X(t-1) ~ M_2 + X(t)
    Y_I = sm.add_constant(Y)
    l2 = sm.OLS(X,Y_I).fit()
    A=l2.params[1:(n+1)]
    C = np.dot(A,B)
    eig_val, eig_vec = np.linalg.eig(C)
    
    #find the smallest eigenvalue
    SE = np.argsort(eig_val)
    
    #Compute the ratio
    ratio = - eig_vec[1,SE[0]]/eig_vec[0,SE[0]]
    return ratio
```

```python
Asset1 = []
Asset2 = []
Weight_list = []
insample_PValue = []
insample_St = []
outsample_PValue = []
outsample_St = []
for comb in comb_list:
    #Save comb
    Asset1.append(comb[0])
    Asset2.append(comb[1])
    
    #Compute Weight
    Weight = CCA_Chou_Ng(comb, sample1)
    Weight_list.append(Weight)
    
    #Compute Portfolio
    Portfolio_in = sample1[comb[0]] - Weight*sample1[comb[1]]
    Portfolio_out = ousample[comb[0]] - Weight*ousample[comb[1]]
    
    #ADF Test
    ADF_in = adfuller(Portfolio_in)
    ADF_out = adfuller(Portfolio_out)
    
    #Stationarity
    insample_PValue.append(ADF_in[1])
    outsample_PValue.append(ADF_out[1])
    if ADF_in[1] < 0.05:
        insample_St.append('Yes')
    else:
        insample_St.append('No')
    
    if ADF_out[1] < 0.05:
        outsample_St.append('Yes')
    else:
        outsample_St.append('No')
    
    ##Combine the result
    di = {'Asset1':Asset1, 
          'Asset2':Asset2, 
          'Weight':Weight_list, 
          'Insample_P':insample_PValue, 
          'Stationarity_in':insample_St, 
          'Outsample_P':outsample_PValue,
          'Stationarity_out':outsample_St
         }
    result_CCA_Chou_Ng = pd.DataFrame(di)
result_CCA_Chou_Ng
```

### Method 3 : CCA_Box_Tiao

```python
def Box_Tiao(comb, data_set):
    
    data = data_set[comb]
    
    #Prepare data
    df = data - data.mean()
    df1 = df.drop(df.index[0])
    df2 = df.shift().dropna()
    
    #Calculate A
    A = (sm.OLS(df1.as_matrix(), df2.as_matrix()).fit()).params
    
    #Calculate Q
    cov = df.cov()
    Q = np.dot(np.dot(np.dot(np.linalg.inv(cov), A), cov),A.T)
    
    #calculate Eigenvector and Eigenvalue
    eig_val, eig_vec = np.linalg.eig(Q)
    
    #find the smallest eigenvalue
    SE = np.argsort(eig_val)
    
    #Compute the ratio
    ratio = - eig_vec[1,SE[1]]/eig_vec[0,SE[1]]
    
    return ratio
```

```python
Asset1 = []
Asset2 = []
Weight_list = []
insample_PValue = []
insample_St = []
outsample_PValue = []
outsample_St = []
for comb in comb_list:
    #Save comb
    Asset1.append(comb[0])
    Asset2.append(comb[1])
    
    #Compute Weight
    Weight = Box_Tiao(comb, sample1)
    Weight_list.append(Weight)
    
    #Compute Portfolio
    Portfolio_in = sample1[comb[0]] - Weight*sample1[comb[1]]
    Portfolio_out = ousample[comb[0]] - Weight*ousample[comb[1]]
    
    #ADF Test
    ADF_in = adfuller(Portfolio_in)
    ADF_out = adfuller(Portfolio_out)
    
    #Stationarity
    insample_PValue.append(ADF_in[1])
    outsample_PValue.append(ADF_out[1])
    if ADF_in[1] < 0.05:
        insample_St.append('Yes')
    else:
        insample_St.append('No')
    
    if ADF_out[1] < 0.05:
        outsample_St.append('Yes')
    else:
        outsample_St.append('No')
    
    ##Combine the result
    di = {'Asset1':Asset1, 
          'Asset2':Asset2, 
          'Weight':Weight_list, 
          'Insample_P':insample_PValue, 
          'Stationarity_in':insample_St, 
          'Outsample_P':outsample_PValue,
          'Stationarity_out':outsample_St
         }
    result_CCA_Box_Tiao = pd.DataFrame(di)
result_CCA_Box_Tiao
```
